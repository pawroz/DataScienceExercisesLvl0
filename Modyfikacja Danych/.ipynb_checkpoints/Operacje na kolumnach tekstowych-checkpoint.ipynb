```python
import pandas as pd
import numpy as np
fortune = pd.read_csv(r'E:\PythonAnacondaScripts\course-files\Fortune_500_2017.csv',usecols =['Rank','Title','Industry','Hqlocation','Sector'])
fortune.head()
```

```python
fortune['Sector'] = fortune['Sector'].str.upper()
```

```python
fortune['Industry'] = fortune['Industry'].str.lower()
```

```python
fortune[fortune['Industry'].str.contains('comp')]
```

```python
City = fortune['Hqlocation'].str.split(' ').str[0]
```

```python
fortune['City'] = City
```

```python
State = fortune['Hqlocation'].str.split(' ').str[1]
```

```python
fortune['State'] = State
```

```python
fortune.head()
```

```python
def BuildShortcut(row):
        result = ''
        industry = row['Industry']
        for i in industry.split(' '):
            result += i[0]
        return result
```

```python
BuildShortcut(fortune.loc[0])
```

```python
fortune['IndustryShort'] = fortune.apply(BuildShortcut,axis=1) 
fortune.head()
```
